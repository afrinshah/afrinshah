```python
gender_count = df['gender'].value_counts()

# Labels and explode values
mylabels = ["Female", "Male", "Non-binary"]
myexplode = [0.2, 0, 0]  # Matches the length of `gender_count`

# Create pie chart
plt.figure(figsize=(5, 5))
plt.pie(gender_count, labels=mylabels, explode=myexplode, autopct='%1.1f%%', shadow=True)
plt.legend(mylabels, title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Gender Distribution")
plt.show()
```

```python
income_by_stream = df.groupby('stream')['monthly_income'].sum()

# Extract data for the donut chart
stream = income_by_stream.index
monthly_income = income_by_stream.values

# Define a color palette (you can customize this)
colors = plt.cm.Paired(range(len(stream)))

# Create donut chart
fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts, autotexts = ax.pie(monthly_income, labels=stream, autopct='%1.1f%%', startangle=140, shadow=True, 
                                  wedgeprops={'width': 0.6}, colors=colors)

# Create a legend
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors]
plt.legend(handles, stream, title='Stream', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add the value in the center of the donut chart
total_income = monthly_income.sum()
ax.text(0, 0, f'Total Income\n${total_income:,.2f}', ha='center', va='center', fontsize=14, color='black', fontweight='bold')

# Set title and show the chart
plt.title('Monthly Income Distribution by Stream')
plt.show()
```

```python
aid_by_stream = df.groupby('stream')['financial_aid'].sum()

# Extract data for the donut chart
stream = aid_by_stream.index
financial_aid = aid_by_stream.values

# Define a color palette (you can customize this)
colors = plt.cm.Paired(range(len(stream)))

# Create donut chart
fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts, autotexts = ax.pie(financial_aid, labels=stream, autopct='%1.1f%%', startangle=140, shadow=True, 
                                  wedgeprops={'width': 0.6}, colors=colors)

# Create a legend
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors]
plt.legend(handles, stream, title='Stream', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add the value in the center of the donut chart
total_aid = financial_aid.sum()
ax.text(0, 0, f'Total Aid\n${total_aid:,.2f}', ha='center', va='center', fontsize=14, color='black', fontweight='bold')

# Set title and show the chart
plt.title('Financial aid Distribution by Stream')
plt.show()
```

```python
food_by_payment = df.groupby('preferred_payment_method')['food expense'].sum()

# Extract data for the bar chart
payment_methods = food_by_payment.index
food_expenses = food_by_payment.values

# Create bar chart
plt.figure(figsize=(6, 6))
bars = plt.bar(payment_methods, food_expenses, color=['blue', 'green', 'orange'])

# Add labels and title
plt.xlabel('Preferred Payment Method')
plt.ylabel('Total Food Expenses')
plt.title('Food Expenses by Preferred Payment Method')

# Create legend handles from the bars
handles = [plt.Rectangle((0, 0), 1, 1, color=bar.get_facecolor()) for bar in bars]

# Add legend
plt.legend(handles, payment_methods, title='Preferred Payment Method', bbox_to_anchor=(1.05, 1), loc='upper left')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
spending_columns = [
    'housing', 'food expense', 'transportation', 'books_supplies',
    'entertainment', 'personal_care', 'technology', 'health_wellness', 'miscellaneous'
]

# Calculate the total spending for each category
total_spending = df[spending_columns].sum()

# Calculate total spending and percentages
grand_total = total_spending.sum()
percentages = (total_spending / grand_total) * 100

# Define a color palette with distinct colors
colors = plt.cm.tab10(np.linspace(0, 1, len(spending_columns)))

# Plotting the bar chart
plt.figure(figsize=(10, 7))
bars = plt.bar(total_spending.index, total_spending.values, color=colors, edgecolor='black')

# Add percentage annotations on top of each bar
for i, (value, percentage) in enumerate(zip(total_spending, percentages)):
    plt.text(
        i, 
        value + grand_total * 0.01,  # Position slightly above the bar
        f"{percentage:.1f}%", 
        ha='center', 
        fontsize=10,
        fontweight='bold'
    )

# Add legend
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors]
plt.legend(handles, spending_columns, title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left', frameon=True)

# Add title and labels
plt.title('Total Spending by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Spending', fontsize=12)

# Adjust the x-axis for better readability
plt.xticks(rotation=45, fontsize=10)

# Adjust layout to fit everything
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='monthly_income', y='tuition', data=df, hue='year_in_college', palette='Set1')

# Title and labels
plt.title('Monthly Income vs. Tuition')
plt.xlabel('Monthly Income')
plt.ylabel('Tuition')

# Adjust legend to be outside the plot
plt.legend(title='Year in College', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent clipping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='year_in_college', y='monthly_income', data=df, hue='year_in_college', palette='Set2')

# Add values on the plot in percentage
for i in range(len(df['year_in_college'].unique())):
    # Getting the median for each box
    median = df[df['year_in_college'] == df['year_in_college'].unique()[i]]['monthly_income'].median()
    # Calculate the median as a percentage of the maximum value
    max_value = df['monthly_income'].max()
    percentage = (median / max_value) * 100
    
    # Add the percentage text to the plot
    ax.text(i, median, f'{percentage:.2f}%', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black', fontweight='bold')

# Title and labels
plt.title('Monthly Income Distribution by Year in College')
plt.xlabel('Year in College')
plt.ylabel('Monthly Income')

# Adjust layout to prevent clipping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Filter the DataFrame to include only the specified columns
columns_to_include = ['stream', 'housing', 'tuition', 'food expense', 
                      'transportation', 'books_supplies', 'entertainment', 'personal_care', 
                      'technology', 'health_wellness', 'miscellaneous']

filtered_df = df[columns_to_include]

# Group by 'year_in_college' and sum the values
stream_summary = filtered_df.groupby('stream').sum()

# Calculate the total for each row and convert values to percentages
percentages = stream_summary.div(stream_summary.sum(axis=1), axis=0) * 100

# Create a horizontal stacked bar chart
percentages.plot(kind='barh', stacked=True, figsize=(12, 7), colormap='tab10', width=0.8)

# Customize the chart
plt.title('Stream Spending across Categories', fontsize=14)
plt.xlabel('Percentage', fontsize=12)
plt.ylabel('Stream', fontsize=12)



# Customize legend
plt.legend(title="Categories", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()  # Adjust layout for better fit

# Show the plot
plt.show()
```

Conclusion:
In conclusion, the analysis of college students' financial behavior reveals significant patterns related to their income 
sources, spending habits, and financial priorities. Factors such as major, housing choices, and financial aid impact how 
students manage their finances, with many prioritizing essential expenses like housing and food. However, discretionary 
spending on entertainment and technology is also prevalent, reflecting lifestyle preferences. Understanding these financial 
behaviors can help colleges, financial institutions, and policymakers design more effective support systems, budget planning 
tools, and financial education programs tailored to students' needs and challenges.
